```python
# Check null values for each table
```

```python
# Collect all loaded tables
all_tables = {
    'customers': customers,
    'sellers': sellers,
    'orders': orders,
    'order_items': order_items,
    'products': products,
    'order_payments': order_payments,
    'order_reviews': order_reviews,
    'geolocation': geolocation,
    'category_translation': category_translation
}
```

```python
print(f"{"TABLE":<30} {"COLUMN":<40} {"NULL COUNT":>12} {"NULL %":>10}")
print("-" * 96)
for name, df in all_tables.items():
    for col in df.columns:
        n = df[col].isnull().sum()
        if n > 0:
            pct = n / len(df) * 100
            print(f"{name:<30} {col:<40} {n:>12,} {pct:>9.2f}%")
```

```python
# Add Metadata Columns - ingested_at, source_filename, batch_id
```

```python
from datetime import datetime, timezone
import uuid
```

```python
ingested_at = datetime.now(timezone.utc)
batch_id = str(uuid.uuid4())

file_map = {
    'customers':            'olist_customers_dataset.csv',
    'sellers':              'olist_sellers_dataset.csv',
    'orders':               'olist_orders_dataset.csv',
    'order_items':          'olist_order_items_dataset.csv',
    'products':             'olist_products_dataset.csv',
    'order_payments':       'olist_order_payments_dataset.csv',
    'order_reviews':        'olist_order_reviews_dataset.csv',
    'geolocation':          'olist_geolocation_dataset.csv',
    'category_translation': 'product_category_name_translation.csv',
}

for name, df in all_tables.items():
    df['ingested_at']     = ingested_at
    df['source_filename'] = file_map[name]
    df['batch_id']        = batch_id
```

```python
import os
from google.cloud import bigquery
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "../gcp-key.json"
```

```python
PROJECT_ID = "coral-idiom-451004-i8"
DATASET    = "raw"

client = bigquery.Client(project=PROJECT_ID)

table_map = {
    'customers':            'raw_customers',
    'sellers':              'raw_sellers',
    'orders':               'raw_orders',
    'order_items':          'raw_order_items',
    'products':             'raw_products',
    'order_payments':       'raw_payments',
    'order_reviews':        'raw_reviews',
    'geolocation':          'raw_geolocation',
    'category_translation': 'raw_product_category_translation',
}

job_config = bigquery.LoadJobConfig(write_disposition="WRITE_TRUNCATE")

for name, df in all_tables.items():
    table_id = f"{PROJECT_ID}.{DATASET}.{table_map[name]}"
    job = client.load_table_from_dataframe(df, table_id, job_config=job_config)
    job.result()
```
